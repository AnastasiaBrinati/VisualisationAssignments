### Challenge 1: Into the Forest of the Howling Sorrows (Grade E)

```python
%matplotlib inline
import numpy as np
from matplotlib import pyplot as mpl
import pandas as pd
```

```python
# Extracting dataframe to analyze the howling sorrows coming from the forest
howling_sorrows = pd.read_excel("DataAssignment1V2.xlsx", sheet_name = "Sorrows")
howling_sorrows = howling_sorrows.rename(columns={"Sorrows Day 1": "Sorrows_Day_1", "Sorrows Day 2": "Sorrows_Day_2","Sorrows Day 3": "Sorrows_Day_3"})

x_labels = howling_sorrows.Time
data_day_1 = howling_sorrows.Sorrows_Day_1
data_day_2 = howling_sorrows.Sorrows_Day_2
data_day_3 = howling_sorrows.Sorrows_Day_3

data_x = np.arange(len(x_labels))
data_y = np.arange(0,340,20)
width = 0.20
fig = mpl.figure(figsize=(14,7))
ax = fig.add_subplot()

ax.bar(data_x - width , data_day_1, width, color = 'red', label = 'Day 1')
ax.bar(data_x, data_day_2, width, color = 'orange', label = 'Day 2')
ax.bar(data_x + width , data_day_3, width, color = 'yellow', label = 'Day 3')

ax.set_xticks(data_x, x_labels)
ax.set_yticks(data_y)
ax.set_title("Howling Sorrows distribution over three nights")
ax.set_xlabel('Night Hours')
ax.set_ylabel('N° of sorrows')
ax.legend()
ax.grid(alpha=0.25)

# Highligthing the best interval to enter the forest
mpl.axvspan(11.70, 12.30, color="blue", alpha=0.2)
```

- **During what interval of time should your party attempt to enter the forest?** 
  Considering that the howling sorrows are a threat for my party, my intention is to enter the forest during the last hour of the night. The data I obtained shows every hour how many sorrows were counted, meaning that the most dangerous
  interval is the first one, and the least one is the last one, right before the sunrise.
- **What trends do you see in the data?**
  The number of sorrows follows the same pattern every day: the rate decreases during the night reaching its minimum in the
  last hour.

### Challenge 3: The Orb of the Eternal Frost (Grade C)

```python
# Obtaining dataframes for each characteristic of each character to obtain the orb

human_vs_orb = pd.read_excel("DataAssignment1V2.xlsx", sheet_name = 'Orb',usecols="A,B,C")
human_vs_orb = human_vs_orb.rename(columns={"Human": "Roll", "Unnamed: 1": "Dexterity_d20","Unnamed: 2": "Intelligence_d12"})
human_vs_orb = human_vs_orb.drop(index=0)
human_vs_orb = human_vs_orb.dropna().reset_index(drop=True)

firbolg_vs_orb = pd.read_excel("DataAssignment1V2.xlsx", sheet_name = 'Orb', usecols="E,F,G")
firbolg_vs_orb = firbolg_vs_orb.rename(columns={"Firbolg": "Roll", "Unnamed: 5": "Dexterity_d8","Unnamed: 6": "Intelligence_d20"})
firbolg_vs_orb = firbolg_vs_orb.drop(index=0)
firbolg_vs_orb = firbolg_vs_orb.dropna().reset_index(drop=True)
```

```python
# Figure 1
# Axes 1&2: Dexterity

fig1 = mpl.figure(figsize=(14,7))
ax = fig1.subplots(nrows=1, ncols=2)

x_rolls = human_vs_orb.Roll
x_labels = ["Human","Firbolg"]

human_dexterity = human_vs_orb.Dexterity_d20
firbolg_dexterity = firbolg_vs_orb.Dexterity_d8

human_intelligence = human_vs_orb.Intelligence_d12
firbolg_intelligence = firbolg_vs_orb.Intelligence_d20

ax[0].set_title('Dexterity')
ax[0].set_ylabel('Dexterity value')
ax[0].set_xlabel('Rolls')
ax[0].set_ylim(0,21)
ax[0].set_xlim(1,11,1)
ax[0].plot(x_rolls, human_dexterity, color='#e75480', label="Human")
ax[0].plot(x_rolls, firbolg_dexterity, color='#5090FF', label="Firbolg")
ax[0].set_yticks(np.arange(len(x_rolls)+1)*2)
ax[0].set_xticks(np.arange(len(x_rolls)+1))
ax[0].grid(alpha=0.5)
max = np.max(firbolg_dexterity)
min = np.min(firbolg_dexterity)
ax[0].fill_between(range(1,11,1), max, min, color='lightblue', alpha = 0.3)
ax[0].legend()

ax[1].set_title('Dexterity')
ax[1].set_ylabel('Dexterity value')
ax[1].set_ylim(0,21)
bp_human = ax[1].boxplot(human_dexterity, positions=[0], patch_artist=True)
bp_firbolg = ax[1].boxplot(firbolg_dexterity, positions=[1], patch_artist=True)
ax[1].set_xticks(np.arange(len(x_labels)), x_labels)
ax[1].set_yticks(np.arange(len(x_rolls)+1)*2)
ax[1].grid(alpha=0.5)

# Coloring the boxplots as before to maintain consistency: pink for the human, lightblue for the firbolg
for patch in bp_human['boxes']:
    patch.set(facecolor="pink")
    mpl.setp(bp_human["medians"], color="red")
for patch in bp_firbolg['boxes']:
    patch.set(facecolor="lightblue")
    mpl.setp(bp_firbolg["medians"], color="red")
ax[1].legend([bp_firbolg["medians"][0]], ['median'], loc='upper right')

fig1.tight_layout()
```

### Challenge 4: The Gelid Palace (Grade B)

```python
# 3D graph of the ice palace

fig = mpl.figure(figsize=(14,7))
ax = fig.add_subplot(projection='3d')

ax.set_ylim(-5,5)
ax.set_xlim(-5,5)
ax.set_zlim(-3,1)

ax.set_yticks(np.arange(-5,6))
ax.set_xticks(np.arange(-5,6))
ax.set_zticks(np.arange(-3,4))

ax.set_ylabel('Y')
ax.set_xlabel('X')
ax.set_zlabel('Z')
ax.set_title("Ice Palace")

x = np.arange(-5,5,0.25)
y = np.arange(-5,5,0.25)

x,y = np.meshgrid(x,y)

ground = -(np.arctan(x*y))*(np.arctan(x*y))
palace = ((np.arctan(x*y))*(np.arcsin(x*y)))+1.7

ax.plot_wireframe(x,y,palace)
ax.plot_wireframe(x,y,ground)

ax.plot_surface(x,y,ground, cmap=mpl.cm.gray)
ax.plot_surface(x,y,palace, cmap=mpl.cm.coolwarm)



# The second figure below is basically the
# same but it has different 
# colors with which I tried to play aroud to have 
# more of an icy-mountain feeling.

fig2 = mpl.figure(figsize=(14,7))
ax2 = fig2.add_subplot(projection='3d')

ax2.set_ylim(-5,5)
ax2.set_xlim(-5,5)

ax2.set_yticks(np.arange(-5,6))
ax2.set_xticks(np.arange(-5,6))

# Z axis adjustments
ax2.set_zlim(-3,1)
ax2.set_zticks(np.arange(-3,5))

ax2.set_ylabel('Y')
ax2.set_xlabel('X')
ax2.set_zlabel('Z')
ax2.set_title("Ice Palace Revisited Colors")

from matplotlib.colors import LinearSegmentedColormap

blues_cmp = LinearSegmentedColormap.from_list("custom", ["#d5896f","#0077b6","#ffffff"])

ax2.plot_surface(x,y,ground, cmap=blues_cmp, antialiased=True)
ax2.plot_surface(x,y,palace, cmap=mpl.cm.Blues_r, antialiased=True)
```

### Challenge 5: The Awakening of Joxdhor (Grade A)

```python
# After generating the data to fill out the tables..
# Obtaining dataframes for each characteristic of each character

#Phase 1: Evasion Phase
evasion = pd.read_excel("DataAssignment1V2.xlsx", sheet_name = 'Joxdhor', usecols="A,B,C,D")
evasion = evasion.rename(columns={"Evade": "Roll", "Unnamed: 1": "Human_Dexterity_d20","Unnamed: 2": "HalfOrc_Dexterity_d12", "Unnamed: 3":"Firbolg_Dexterity_d8"})
evasion = evasion.drop(index=1)
evasion = evasion.dropna().reset_index(drop=True)

#Phase 2: Resist Phase
resistance = pd.read_excel("DataAssignment1V2.xlsx", sheet_name = 'Joxdhor', usecols="F,G,H,I,J,K,L")
resistance = resistance.rename(columns={"Resist": "Roll", "Unnamed: 6": "Human_Endurance_d10","Unnamed: 7": "Human_Intelligence_d12", "Unnamed: 8":"HalfOrc_Endurance_d10", "Unnamed: 9": "HalfOrc_Intelligence_d6","Unnamed: 10": "Firbolg_Endurance_d6", "Unnamed: 11": "Firbolg_Intelligence_d20"})
resistance = resistance.drop(index=1)
resistance = resistance.dropna().reset_index(drop=True)

#Phase 3: Attack Phase
attack = pd.read_excel("DataAssignment1V2.xlsx", sheet_name = 'Joxdhor', usecols="N,O,P,Q")
attack = attack.rename(columns={"Attack": "Roll", "Unnamed: 14": "Human_Dexterity_d20","Unnamed: 15": "HalfOrc_Strenght_d20", "Unnamed: 16":"Firbolg_Intelligence_d20"})
attack = attack.drop(index=1)
attack = attack.dropna().reset_index(drop=True)
```

```python
# Fight with Joxdhor

fig = mpl.figure(figsize=(14,12))
ax = fig.subplots(nrows=2, ncols=2)
rolls = evasion.Roll
data_y = np.arange(0,21,1)
data_x = np.arange(len(rolls))
width = 0.20


# Axes 1: Evasion Phase

human_dexterity = evasion.Human_Dexterity_d20
halforc_dexterity = evasion.HalfOrc_Dexterity_d12
firbolg_dexterity = evasion.Firbolg_Dexterity_d8

# Highlighting value to evade Joxdhor's attack
dexterity_threshold = 8
ax[0][0].axhline(y=dexterity_threshold, color='red', label='Dexterity Threshold')

ax[0][0].bar(data_x - width, halforc_dexterity, width, color = 'lightgreen', label = 'Half-Orc')
ax[0][0].bar(data_x, human_dexterity, width, color = 'pink', label = 'Human')
ax[0][0].bar(data_x + width , firbolg_dexterity, width, color = 'lightblue', label = 'Firbolg')

ax[0][0].set_xticks(data_x, rolls)
ax[0][0].set_yticks(data_y)
ax[0][0].set_title("Evasion Phase: Dexterity")
ax[0][0].set_xlabel('Rolls')
ax[0][0].set_ylabel('Dexterity value')
ax[0][0].legend(loc='upper left')
ax[0][0].grid(alpha=0.25)


# Axes 2: Resistance Phase

human_endurance = resistance.Human_Endurance_d10
halforc_endurance = resistance.HalfOrc_Endurance_d10
firbolg_endurance = resistance.Firbolg_Endurance_d6

# Highlighting value to resist Joxdhor's attack
endurance_threshold = 7
ax[0][1].axhline(y=endurance_threshold, color='red', label='Endurance Threshold')

ax[0][1].bar(data_x - width, halforc_endurance, width, color = 'lightgreen', label = 'Half-Orc')
ax[0][1].bar(data_x, human_endurance, width, color = 'pink', label = 'Human')
ax[0][1].bar(data_x + width , firbolg_endurance, width, color = 'lightblue', label = 'Firbolg')

ax[0][1].set_xticks(data_x, rolls)
ax[0][1].set_yticks(data_y)
ax[0][1].set_title("Resistance Phase: Endurance")
ax[0][1].set_xlabel('Rolls')
ax[0][1].set_ylabel('Endurance value')
ax[0][1].legend(loc='upper right')
ax[0][1].grid(alpha=0.25)


# Axes 3: Resistance Phase

human_intelligence = resistance.Human_Intelligence_d12
halforc_intelligence = resistance.HalfOrc_Intelligence_d6
firbolg_intelligence = resistance.Firbolg_Intelligence_d20

# Highlighting value to resist Joxdhor's attack
intelligence_threshold = 7
ax[1][0].axhline(y=intelligence_threshold, color='red', label='Intelligence Threshold')

ax[1][0].bar(data_x - width, halforc_intelligence, width, color = 'lightgreen', label = 'Half-Orc')
ax[1][0].bar(data_x, human_intelligence, width, color = 'pink', label = 'Human')
ax[1][0].bar(data_x + width , firbolg_intelligence, width, color = 'lightblue', label = 'Firbolg')

ax[1][0].set_xticks(data_x, rolls)
ax[1][0].set_yticks(data_y)
ax[1][0].set_title("Resistance Phase: Intelligence")
ax[1][0].set_xlabel('Rolls')
ax[1][0].set_ylabel('Intelligence value')
ax[1][0].legend(loc='upper right')
ax[1][0].grid(alpha=0.25)


# Axes 4: Attack Phase

human_endurance = attack.Human_Dexterity_d20
halforc_endurance = attack.HalfOrc_Strenght_d20
firbolg_endurance = attack.Firbolg_Intelligence_d20

# Highlighting value to attack Joxdhor
attack_threshold = 15
ax[1][1].axhline(y=attack_threshold, color='red', label='Attack Threshold')

ax[1][1].bar(data_x - width, halforc_endurance, width, color = 'lightgreen', label = 'Half-Orc:Strenght')
ax[1][1].bar(data_x, human_endurance, width, color = 'pink', label = 'Human:Dexterity')
ax[1][1].bar(data_x + width , firbolg_endurance, width, color = 'lightblue', label = 'Firbolg:Intelligence')

ax[1][1].set_xticks(data_x, rolls)
ax[1][1].set_yticks(data_y)
ax[1][1].set_title("Attack Phase: Best Damage Attribute")
ax[1][1].set_xlabel('Rolls')
ax[1][1].set_ylabel('Best Attribute Value')
ax[1][1].legend(loc='upper right')
ax[1][1].grid(alpha=0.25)

fig.tight_layout()
```

From the first axes emerges that only the human has some chance to survive the initial attack, while the half-orc may barely escape. Unfortunatly the firbolg would have no chances to evade.

- **Which characters are most probably able to resist Joxdhor damage if hit?**
Thanks to her dexterity and her intelligence the human would manage to resist Joxdhor, as shown in axes 2 (top right) and 3 (bottom left); the firbolg could find a way to defend himself using his intelligence, but since he doesn't have enough endurance, if his plan where to fail, he couldn't endure the attack.
